### Una descripción formal de la agrupamiento (clustering)

La _'teoría de perturbación' cosmológica es una herramienta que utilizamos para describir la relación entre el 'agrupamiento' (cantidades relativas entre diferentes grandes volúmenes) de _toda_ la materia y las galaxias.  La ecuación que relaciona estas cantidades puede parecer confusa a primera vista, pero en realidad no es más que una simple descripción de esta idea, escrita como un polinomio con (quizás más de una) variable independiente. 

Quizás tengas experiencia trabajando con "polinomios" que tienen una "variable independiente", por ejemplo, la ecuación de una recta $f(x)= mx + b$, es un polinomio de una variable (la variable x) de entrada, por ejemplo, el número de clientes atendidos, $f(x)$, para un número determinado de personal, $x$.  Del mismo modo, la ecuación de una función cuadrática $f(x) = ax^2 + bx + c$, podría describir un comportamiento más complejo para la misma situación. 

Otros pueden haber trabajado con ecuaciones que poseen dos variables independientes, como la ecuación de un círculo: $x^2+y^2=r^2$.  Aquí $(x,y)$ describe una posición en el plano, $r$ es el radio de un círculo, y los puntos en el plano que satisfacen esta _restricción_ dibujarían un círculo.

Si estás acostumbrado a ver estas ecuaciones, la ecuación de la agrupación no te parecerá tan mala.  Si no, pronto te acostumbrarás a ella.  

Sin más preámbulos, presento una de las muchas ecuaciones que utilizo para describir el **agrupamiento galáctico**:

$\delta_g = b_1\delta_m + \frac{b_2}{2}\left[\delta^2_m-\langle \delta^2_m \rangle\right] +\frac{b_s}{2}\left[s^2-\langle s^2 \rangle\right] + \ldots$

Vamos a tomar un minuto para hablar de lo que cada pedazo de esto significa, comenzando con nuestro primer fragmento: 

$\delta_g = b_1\delta_m + \ldots$

La letra griega delta ($\delta$) representa una _sobredensidad_ - una medida de cómo la densidad relativa del volumen que estamos viendo se compara con la densidad media del Universo.  El subíndice $g$ describe esto para las galaxias y $m$ hace lo mismo para la materia total.  Pongamos un ejemplo, digamos que nuestro número medio de galaxias en un volumen puede representarse por $\bar n_g$, un volumen con $\delta_g = \left (n / \bar n\right ) - 1 =2$, tendría $\times 3$ el número de galaxias de lo normal, ¡por definición!  Mientras que $b_1$ es un número desconocido (o _'coeficiente de sesgo'_) que permite una flexibilidad suficiente para describir el Universo que vemos.  Para nuestro ejemplo, si $\delta_g = 2$ y $b_1=4$, entonces tenemos $\delta_m = 0,5$, o sea, el mismo volumen con $\times 1.5$ la cantidad de _materia_ que la media, a pesar de que haya un número mucho mayor de _galaxias_ que la media.  La formación de galaxias es compleja, así que simplemente escribimos un modelo flexible de cómo se forman las galaxias en el Universo y encontramos el (conjunto de coeficientes de sesgo) que mejor describe nuestros datos.


Tomemos un modelo unidimensional de juguete del Universo en el que Mario, ¡o tú!, sólo puedan caminar en línea recta:

```python
# Distancia desde el punto inicial, la derecha es positiva, la izquierda es negativa.   
displacements = np.arange(-10, 10., 1)
```

Donde las fluctuaciones de la densidad de materia de un punto a otro están bien descritas por una distribución normal, 

```python
delta_m = np.random.normal(loc=0.0, scale=1.0, size=len(displacements))
```

```python
pl.plot(displacements, delta_m, c='k', lw=0.2)

pl.xlabel('Distancia desde el punto inicial.')
pl.ylabel('$\delta_m$')
```

Vemos que, a nuestra izquierda, $\delta_m = -1,5$, lo que representa un gran _vacío_, con relativamente muy poca masa.  Mientras que a nuestra derecha, el gran pico sería un denso cúmulo de masa en nuestro Universo de juguete.

¿Qué puede describir nuestro primer modelo simple de formación de galaxias?  Tomemos un rango para los valores $b_1$ _pueden_ ser,   

```python
b1s = np.arange(0.5, 2.5, 0.5)
b1s = b1s[b1s != 1.0]
```

```python
pl.plot(displacements, delta_m, c='k', lw=0.5)

for b1 in b1s:
    pl.plot(displacements, b1 * delta_m, lw=0.2, label='$b_1: {:.1f}$'.format(b1))

pl.xlabel('Distancia desde el punto inicial.')
pl.ylabel('$\delta_m$')
pl.legend(frameon=False)
```

Como se anunció, ¡tenemos una serie de Universos en los que las galaxias se forman de forma diferente!  Cada color muestra un Universo de este tipo, en el que el número de galaxias varía, a pesar de que la cantidad de materia de fondo es la misma entre ellos.

Un modelo más desarrollado

$\delta_g = b_1\delta_m + \frac{b_2}{2}\left[\delta^2_m-\langle \delta^2_m \rangle\right]$ 

describe un comportamiento más complejo entre estas mismas densidades relativas.  En nuestro primer modelo, ($\delta_g / \delta_m$) es siempre la constante $b_1$, mientras que esto se flexibiliza en nuestro segundo modelo.  A veces, esto se adopta para describir mejor volúmenes más extremos en el Universo, en una región densa particular del Universo cuando $\delta_m$ es un número grande, veríamos un comportamiento muy diferente para nuestro primer modelo "lineal".  Veamos un ejemplo, 


```python
b2 = 1.5
```

```python
pl.plot(displacements, delta_m, c='k', lw=0.5)

for b1 in b1s:
    pl.plot(displacements, b1 * delta_m + 0.5 * b2 * (delta_m**2. - np.mean(delta_m**2.)), lw=0.2, label='$b_1: {:.1f}$'.format(b1))

pl.xlabel('Distancia desde el punto inicial.')
pl.ylabel('$\delta_m$')
pl.legend(frameon=False)
```

Para volúmenes cercanos a la densidad media $\delta_m \simeq 0$ vemos muy poca diferencia respecto al primer modelo, ¡mientras que el número de galaxias en nuestro cúmulo denso se ha duplicado!  No hemos visto antes $\langle \delta_m \rangle$, pero es una idea sencilla: representa la media de $\delta_m$ sobre todos los volúmenes.  Como hemos definido que se trata de fluctuaciones relativas, entonces $\langle \delta_m \rangle = \langle n \rangle / \bar{n} - 1$, pero $\langle n \rangle$ es nuestra densidad media de $\bar n$, por lo que $\langle \delta_m \rangle$ debe ser cero.  Aplicando la misma idea a $\langle \delta_m \rangle$ significa que nuestro modelo debe contener este extraño término $\langle \delta_m^2 \rangle$, ¡o no será físico!       

Nuestro último, y más avanzado, conjunto de modelos de Universos son descritos por:

$\delta_g = b_1\delta_m + \frac{b_2}{2}\left[\delta^2_m-\langle \delta^2_m \rangle\right] +\frac{b_s}{2}\left[s^2-\langle s^2 \rangle\right] + \ldots$

En este caso, el número de galaxias en un volumen dado no está determinado únicamente por la cantidad de materia, $\delta_m$.  También puede influir la fuerza de marea en ese volumen.  Por ejemplo, la fuerza de marea sobre una protogalaxia que intenta formarse puede ser un "cizallamiento diferencial", un término elegante para describir una idea obvia de que la galaxia puede sentir una fuerza más fuerte que tira en una dirección hacia un lado, en lugar de en el otro.  Esto puede hacer que la galaxia tenga más o menos posibilidades de formarse.  No lo sabemos, pero es posible.  Podemos entonces incluir esta idea pero añadiendo una inocente $s$, un "tensor" de marea, que describe justamente esta aceleración gravitatoria de dos trozos de materia vecinos.

### ¿Qué complejidad debe tener nuestro modelo de Universo?

La única cantidad que podemos medir directamente y comparar con nuestro modelo es $\delta_g$.  ¿Cómo sabemos entonces qué efectos diferentes deben aparecer en el lado derecho de la ecuación?  Una descripción completa requiere responder a dos preguntas distintas: ¿Cuáles son las propiedades, o "campos" -una cantidad física como la temperatura definida en cada punto del espacio-, que nos deben importar?  En el ejemplo anterior, serían $\delta_m$ y $s$, pero ¿hay otros?  ¿De qué manera se pueden incluir esos campos y seguir siendo físicos?  Tuvimos $\delta_m$ sin problema, pero tuvimos que limitarnos a $\delta_m^2 - \langle \delta_m^2 \rangle$ para que fuera una descripción válida de la naturaleza.  

Un principio fundamental de la teoría de perturbaciones cosmológicas (¡hoy en día se podría diseñar una versión mejor!) es que la gravedad es la única fuerza de largo alcance que nos interesa. Esto significa que cada campo que incluyamos debe describir la gravedad de una forma u otra.  Un tipo de campo que tiene sentido incluir es la sobredensidad de la materia. Podríamos tratar la materia normal y la materia oscura con campos diferentes, pero es más común ponerlos juntos. Otro tipo de campo que podríamos incluir es el relacionado con los cambios en la velocidad de nuestra materia, ya que la aceleración gravitatoria provoca esos cambios.  No incluimos una dependencia de la formación de galaxias en relación con la velocidad de la materia oscura en ese punto debido a la relatividad especial, cuyo fundamento es que las leyes de la naturaleza son las mismas independientemente de la velocidad a la que se viaje.  Existen otros campos y podrían incluirse, ¡pero nos detendremos aquí!

Una vez que tenemos los campos que nos interesan podemos pensar en la mejor manera de incluirlos para hacer un modelo del Universo más realista.  **Las simetrías** nos dan la respuesta a esta pregunta.  Intenta esto: Coge un papel y dibuja un cuadrado.  Ahora gira ese cuadrado 90 grados. Fíjate en que el cuadrado tiene el mismo aspecto que antes de girarlo. Llamamos a esta rotación simetría de un cuadrado, ya que su aspecto es similar después de esta rotación.  En cambio, si ahora giras el cuadrado por algún otro ángulo que no sea múltiplo de 90 grados, verás que el cuadrado tiene un aspecto diferente. Este tipo de rotaciones no son simetrías del cuadrado.

¿Qué simetrías posee esta obra de Escher?  ¿Y si la imprimimos en blanco y negro?  La obra de Escher comparte mucha filosofía con la Relatividad General de Einstein, pero no trataremos eso aquí.

Cuando pensamos en cómo construir nuestra ecuación de agrupación, o modelo de Universo, tenemos que asegurarnos de que todo lo que añadimos tiene las simetrías correctas del problema en cuestión, de lo contrario no es una descripción válida.  Tanto Copérnico como Einstein creían que la física entendía que todos eran iguales, y nuestro modelo debería satisfacer también su punto de vista.  Al imponer simetrías en nuestros modelos, sólo estamos estipulando que las leyes del Universo deben ser las mismas en un lugar, o a una velocidad, que en cualquier otro.  

En este ejemplo, esta es la razón por la que nuestro modelo incluye el cuadrado de $s^2$, pero no $s$; si incluyéramos $s$, no todos los observadores estarían de acuerdo.
